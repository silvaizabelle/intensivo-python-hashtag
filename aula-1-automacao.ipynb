```python
# Passo 4: Importar a base de vendas para Python

# Instalar biblioteca para análise de dados
# pandas
# numpy
# openxL
# Se estiver usando o Jupyter, não precisa instalar

import pandas as pd # Dar apelido para a biblioteca, assim não precisa escrever o nome inteiro

tabela = pd.read_excel(r"C:\Users\bruno\Downloads/Vendas - Dez.xlsx") 
# Informar diretório, onde está o arquivo
# O r antes do diretório indica que é pro Python ler o texto normal, sem identificar os caracteres especiais
# Armazenar a informação em alguma variável
# A nossa tabela será essa que o pandas carregará
# A tabela recebe o valor que está colocado à direita

display(tabela)
# Display apresenta o formato de tabela comoconhecemos
```

```python
# Passo 5: Calcular faturamento e quantidade de produtos vendidos

# Colchetes [] serve para selecionar as colunas numa base de dados
# Os colchetes selecionando a coluna é da biblioteca do Pandas

faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()

# .sum serve para somar

# Visualizar resultados
# display(faturamento)
# display(quantidade)
```
